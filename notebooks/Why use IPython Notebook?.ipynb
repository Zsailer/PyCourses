# Why use the IPython Notebook for these lessons?

Alright, I'll admit it. I'm a bias since I've been a developer for the project in the past. 

But hear me out for a minute.

You've already made the decision to learn some Python, a fantastic choice. IPython Notebook simply takes Python off the commandline/terminal and let's you create code **interactively**. 

You can run and debug (a buzzword for "fix") your code instantaneously. There's no need to save a script, close your VI window, and run:

```>>> python <myscript>```

on the command line. 

If you've worked in *Matlab* or *Mathematica* before, you've experienced this behavior. They have *cells* to break up your code and run small chunks rather than whole scripts. That's super useful if you want to quickly check your code.

So why not use R, Matlab, or Mathematica? I mean, I'm heard Wolfram Alpha can tell you the answer to the universe... 

I'm not claiming these languages aren't good. However, they were created for specific purposes and won't easily deviate from them. R is great for interactive statistical analysis, but is very unintuitive when trying to develop larger projects  and analysis pipelines. Matlab cannot be beat when it comes to mathematical matrix manipulation in a few lines of code. However, anything beyond this purpose becomes very nasty. Mathematica works well for interactive analysis of pure math but isn't a real language. It cannot be used for general programming.

To a scientist, the strength of Python can be summarized as so:

**It's the most intuitive, readable, and real programming language out there for scientists.**

Paired with the Notebook and other easily accessible open source packages, it can do everything these other languages can do while keeping its status as a real language.
