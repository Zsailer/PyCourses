## Python

If you've worked in *Matlab* or *Mathematica* before, you've experienced this behavior. They have *cells* to break up your code and run small chunks rather than whole scripts. That's super useful if you want to quickly check your code.

So why not use R, Matlab, or Mathematica? I mean, I'm heard Wolfram Alpha can tell you the answer to the universe... 

I'm not claiming these languages aren't good. However, they were created for specific purposes and won't easily deviate from them. R can be great for interactive statistical analysis, but is very unintuitive when trying to develop larger projects  and analysis pipelines. Matlab cannot be beat when it comes to mathematical matrix manipulation in a few lines of code. However, anything beyond this purpose becomes very nasty (not to mention slow). Mathematica works well for interactive analysis of pure math but isn't a real language. It cannot be used for general programming.

To a scientist, the strength of Python can be summarized as so:

**It's the most intuitive, readable, and real programming language out there for scientists.**

Paired with the Notebook and other easily accessible open source packages, it can do everything these other languages can do while keeping its status as a "real" language. Specifically, there are many frameworks available that bridge Python and Javascript, making the Python accessible to the Browser (IPython is a explicit example of this). This is fantastic! In the age of app development and data sharing, the browser is the leading platform for development. Also, if you need speed, Python easily dips in the the the language C (the typical benchmark for speed) via Cython and extensions.
